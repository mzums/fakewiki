```python
import torch
```

```python
# standard deviation grows inside the residual connection
x = torch.zeros(768)
n = 100
for i in range(n):
    x += torch.randn(768)

print(x.std())
```

```python
# so we have to scale every contribution
x = torch.zeros(768)
n = 100
for i in range(n):
    x += n**-0.5 * torch.randn(768)

print(x.std())
```
